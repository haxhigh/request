```python
import pandas as pd

df = pd.read_csv("Merged_data_clean.csv")
df
```

```python
Solar_mass = df["mass"]
Solar_mass
Si_mass = []
for i in Solar_mass:
    i = float(i) * 1.989e+30
    Si_mass.append(i)
Si_mass
```

```python
Solar_radius = df["radius"]
Solar_radius
si_radius = []
for i in Solar_radius:
    i = float(i) * 6.957e+8
    si_radius.append(i)
si_radius
```

```python
planet_names = df["name"]
planet_names
```

```python
planet_names = df["name"]

Gravity = []
def calculateGravity():
    for index,name in enumerate(planet_names):
        gravity = (float(Solar_mass[index])*5.972e+24) / (float(Solar_radius[index])*float(Solar_radius[index])*6371000*6371000)*6.674e-11
        Gravity.append(gravity)
calculateGravity()
GravityDF = pd.DataFrame(list(zip(Gravity)))
```

```python
merged = pd.merge(GravityDF,df,how="right")
merged
```
